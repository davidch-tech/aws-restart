# 2 Working with Numeric Data Types

```python
# Calculator

def operation_menu(operation):
    while True:
        if operation == "1":
            result = add(num1, num2)
            break
        elif operation == "2":
            result = subtract(num1, num2)
            break
        elif operation == "3":
            result = multiply(num1, num2)
            break
        elif operation == "4":
            result = divide(num1, num2)
            break
        else:
            print("Invalid operation selected. Please choose a valid operation.")
            operation = input("Choose an operation (1:add, 2:subtract, 3:multiply, 4:divide): ")
    return result

def get_integer(prompt):
    """Ask the user for input and validate it is a whole integer."""
    while True:
        raw = input(prompt)
        # TODO(human): validate that `raw` is an integer and return it.
        # If it's not valid, print an error and continue the loop.
        try:
            if "." not in raw:
                value = int(raw)
                print(f"Received valid integer: {value}")
            else:
                value = float(raw)
                print(f"Received valid float: {value}")
            return value
        except ValueError:
            print(f"Invalid input '{raw}'. Please enter a valid number.")

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    else:
        print("Error: Division by zero is not allowed.")
        return None

num1 = get_integer("Enter the first whole number: ")
num2 = get_integer("Enter the second whole number: ")

operation = input("Choose an operation (1:add, 2:subtract, 3:multiply, 4:divide): ")

result = operation_menu(operation)

if result is not None:
    print(f"The result is: {result}")
```

# 6 Working with Composite Data Types

```python
import csv
import copy

myVehicle = {
    "vin" : "<empty>",
    "make" : "<empty>" ,
    "model" : "<empty>" ,
    "year" : 0,
    "range" : 0,
    "topSpeed" : 0,
    "zeroSixty" : 0.0,
    "mileage" : 0
}

for key, value in myVehicle.items():
    print(f"{key} : {value}".format(key, value))

myInventoryList = []

with open('car_fleet.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')  
    lineCount = 0  
    for row in csvReader:
        if lineCount == 0:
            print(f'Column names are: {", ".join(row)}')  
            lineCount += 1  
        else:  
            print(f'vin: {row[0]} make: {row[1]}, model: {row[2]}, year: {row[3]}, range: {row[4]}, topSpeed: {row[5]}, zeroSixty: {row[6]}, mileage: {row[7]}')  
            currentVehicle = copy.deepcopy(myVehicle)  
            currentVehicle["vin"] = row[0]  
            currentVehicle["make"] = row[1]  
            currentVehicle["model"] = row[2]  
            currentVehicle["year"] = row[3]  
            currentVehicle["range"] = row[4]  
            currentVehicle["topSpeed"] = row[5]  
            currentVehicle["zeroSixty"] = row[6]  
            currentVehicle["mileage"] = row[7]  
            myInventoryList.append(currentVehicle)  
            lineCount += 1  
    print(f'Processed {lineCount} lines.')
```

```python
currentVehicle = copy.deepcopy(myVehicle)
```

```python
for myCarProperties in myInventoryList:
    for key, value in myCarProperties.items():
        print("{} : {}".format(key,value))
        print("-----")
```

```python
userReply = input("Would you like to buy stamps, buy an envelope, or make a copy? (Enter stamps, envelope, or copy) ")
if userReply == "stamps":
    print("We have many stamp designs to choose from.")
elif userReply == "envelope":
    print("We have many envelope sizes to choose from.")
elif userReply == "copy":
    copies = input("How many copies would you like? (Enter a number) ")
    print("Here are {} copies.".format(copies))
else:
    print("Thank you, please come again.")
```
