# 2 Working with Numeric Data Types

```python
# Calculator

def operation_menu(operation):
    while True:
        if operation == "1":
            result = add(num1, num2)
            break
        elif operation == "2":
            result = subtract(num1, num2)
            break
        elif operation == "3":
            result = multiply(num1, num2)
            break
        elif operation == "4":
            result = divide(num1, num2)
            break
        else:
            print("Invalid operation selected. Please choose a valid operation.")
            operation = input("Choose an operation (1:add, 2:subtract, 3:multiply, 4:divide): ")
    return result

def get_integer(prompt):
    """Ask the user for input and validate it is a whole integer."""
    while True:
        raw = input(prompt)
        # TODO(human): validate that `raw` is an integer and return it.
        # If it's not valid, print an error and continue the loop.
        try:
            if "." not in raw:
                value = int(raw)
                print(f"Received valid integer: {value}")
            else:
                value = float(raw)
                print(f"Received valid float: {value}")
            return value
        except ValueError:
            print(f"Invalid input '{raw}'. Please enter a valid number.")

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    else:
        print("Error: Division by zero is not allowed.")
        return None

num1 = get_integer("Enter the first whole number: ")
num2 = get_integer("Enter the second whole number: ")

operation = input("Choose an operation (1:add, 2:subtract, 3:multiply, 4:divide): ")

result = operation_menu(operation)

if result is not None:
    print(f"The result is: {result}")
```
